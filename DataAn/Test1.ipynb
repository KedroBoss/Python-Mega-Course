```python
from geopy.geocoders import Nominatim

nom = Nominatim()
```

```python
n=nom.geocode("3995 23rd St, San Francisco, CA 94114")
```

```python
import pandas
geo_frame = pandas.read_json('supermarkets.json')
geo_frame
```

```python
geo_frame["Address"] = geo_frame["Address"]+', '+geo_frame['City']+', '+geo_frame['State']+', '+geo_frame['Country']
geo_frame
```

```python
geo_frame['Coordinate'] = geo_frame["Address"].apply(nom.geocode)
geo_frame
```

```python
geo_frame['Latitude'] = geo_frame['Coordinate'].apply(lambda x: x.latitude if x != None else None)
geo_frame['Longitude'] = geo_frame['Coordinate'].apply(lambda x: x.longitude if x != None else None)
geo_frame
```
